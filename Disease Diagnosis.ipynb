```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import roc_auc_score
```

Random Forest Classifier for predicting without symptoms (based on supplementary features)

```python
X = df[['Age', 'Heart_Rate_bpm', 'Body_Temperature_C', 'Oxygen_Saturation_%']]
y = df['Diagnosis']
```

```python
X.shape
```

# **Naive Bayes Classifier**

```python
#1. Naive Bayes without symptoms (using only supplementary features)

from sklearn.naive_bayes import GaussianNB

X = df[['Age', 'Heart_Rate_bpm', 'Body_Temperature_C', 'Oxygen_Saturation_%']]
y = df['Diagnosis']

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

nb_no_symptoms = GaussianNB()
nb_no_symptoms.fit(X_train, y_train)

nb_predictions = nb_no_symptoms.predict(X_test)
nb_proba = nb_no_symptoms.predict_proba(X_test)

print("\nNaive Bayes Classification Report (without symptoms):")
print(classification_report(y_test, nb_predictions, target_names=le.classes_))
print(f"Naive Bayes Accuracy (without symptoms): {accuracy_score(y_test, nb_predictions):.2f}")

auc_score = roc_auc_score(y_test, nb_proba, multi_class='ovr', average='macro')
print(f"Naive Bayes AUC-ROC Score (macro avg): {auc_score:.2f}")

scores = cross_val_score(nb_no_symptoms, X, y_encoded, cv=5, scoring='accuracy')
print("Cross-Validation Accuracy Scores:", scores)
print("Mean Accuracy:", scores.mean())
```

```python
#2. Naive Bayes with only symptoms

X = df.drop(['Gender','Diagnosis', 'Heart_Rate_bpm','Body_Temperature_C','Oxygen_Saturation_%',
             'Age','Treatment_Plan','Blood_Pressure_mmHg','Severity','Patient_ID'], axis=1)
y = df['Diagnosis']

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

nb_symptoms = GaussianNB()
nb_symptoms.fit(X_train, y_train)

nb_predictions = nb_symptoms.predict(X_test)
nb_proba = nb_symptoms.predict_proba(X_test)

print("\nNaive Bayes Classification Report (using only symptoms):")
print(classification_report(y_test, nb_predictions, target_names=le.classes_))
print(f"Naive Bayes Accuracy (using only symptoms): {accuracy_score(y_test, nb_predictions):.2f}")

auc_score = roc_auc_score(y_test, nb_proba, multi_class='ovr', average='macro')
print(f"Naive Bayes AUC-ROC Score (macro avg): {auc_score:.2f}")

scores = cross_val_score(nb_symptoms, X, y_encoded, cv=5, scoring='accuracy')
print("Cross-Validation Accuracy Scores:", scores)
print("Mean Accuracy:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()

param_grid = {
    'var_smoothing': np.logspace(-12, -6, 20)
}

# Run grid search with cross-validation
grid_search = GridSearchCV(nb_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X, y_encoded)

# Best parameter and score
print("Best var_smoothing:", grid_search.best_params_['var_smoothing'])
print("Best cross-val accuracy:", grid_search.best_score_)

# Use best model
best_nb_model = grid_search.best_estimator_

# Evaluate on test set
best_nb_predictions = best_nb_model.predict(X_test)
best_nb_proba = best_nb_model.predict_proba(X_test)

print("\nOptimized Naive Bayes Classification Report (using only symptoms):")
print(classification_report(y_test, best_nb_predictions, target_names=le.classes_))
print(f"Optimized Naive Bayes Accuracy: {accuracy_score(y_test, best_nb_predictions):.2f}")

auc_score = roc_auc_score(y_test, best_nb_proba, multi_class='ovr', average='macro')
print(f"Optimized Naive Bayes AUC-ROC Score (macro avg): {auc_score:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score

# Logistic Regression without Symptoms (using the same features as the first Random Forest)
X = df[['Age', 'Heart_Rate_bpm', 'Body_Temperature_C', 'Oxygen_Saturation_%']]
y = df['Diagnosis']

# Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

logreg_predictions = logreg.predict(X_test)
logreg_proba = logreg.predict_proba(X_test)

print("Logistic Regression Classification Report (without symptoms):")
print(classification_report(y_test, logreg_predictions, target_names=le.classes_))
print(f"Logistic Regression Accuracy (without symptoms): {accuracy_score(y_test, logreg_predictions):.2f}")

auc_score = roc_auc_score(y_test, logreg_proba, multi_class='ovr', average='macro')
print(f"Logistic Regression AUC-ROC Score (macro avg): {auc_score:.2f}")

scores = cross_val_score(logreg, X, y_encoded, cv=10, scoring='accuracy')
print("Cross-Validation Accuracy Scores:", scores)
print("Mean Accuracy:", scores.mean())
```

# SVM

```python
# 1. SVM without symptoms (using only supplementary features)
X = df[['Age', 'Heart_Rate_bpm', 'Body_Temperature_C', 'Oxygen_Saturation_%']]
y = df['Diagnosis']

le = LabelEncoder()
y_encoded = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

svm_no_symptoms = make_pipeline(StandardScaler(), SVC(kernel='rbf', probability=True))
svm_no_symptoms.fit(X_train, y_train)

svm_predictions = svm_no_symptoms.predict(X_test)
svm_proba = svm_no_symptoms.predict_proba(X_test)

print("\nSVM Classification Report (without symptoms):")
print(classification_report(y_test, svm_predictions, target_names=le.classes_))
print(f"SVM Accuracy (without symptoms): {accuracy_score(y_test, svm_predictions):.2f}")


auc_score = roc_auc_score(y_test, svm_proba, multi_class='ovr', average='macro')
print(f"SVM AUC-ROC Score (macro avg): {auc_score:.2f}")


scores = cross_val_score(svm_no_symptoms, X, y_encoded, cv=5, scoring='accuracy')
print("Cross-Validation Accuracy Scores:", scores)
print("Mean Accuracy:", scores.mean())
```

# Decision Tree

```python
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

*To handle class imbalance in the dataset, the class_weight='balanced' parameter was used while training the Decision Tree classifiers. This dynamically adjusts the weight of each class inversely proportional to its frequency, ensuring that minority classes are not overlooked during model learning.*

## 1. Decision Tree without Symptoms

```python

X = df[['Age', 'Heart_Rate_bpm', 'Body_Temperature_C', 'Oxygen_Saturation_%']]
y = df['Diagnosis']
le = LabelEncoder()
y_encoded = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
# Gini Criterion
print("\nGini Criterion:")
dt_gini = DecisionTreeClassifier(criterion='gini', class_weight='balanced', random_state=101)
dt_gini.fit(X_train, y_train)
y_pred = dt_gini.predict(X_test)
y_proba = dt_gini.predict_proba(X_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

print("AUC-ROC Score:", roc_auc_score(y_test, y_proba, multi_class='ovr'))

cv_scores = cross_val_score(dt_gini, X, y_encoded, cv=5)
print("\nCross-Validation Scores:", cv_scores)
print("Mean CV Accuracy:", np.mean(cv_scores))
print("Train Accuracy:", dt_gini.score(X_train, y_train))
print("Test Accuracy:", dt_gini.score(X_test, y_test))
```

```python
# Entropy Criterion
print("\nEntropy Criterion:")
dt_entropy = DecisionTreeClassifier(criterion='entropy', class_weight='balanced', random_state=101)
dt_entropy.fit(X_train, y_train)
y_pred = dt_entropy.predict(X_test)
y_proba = dt_entropy.predict_proba(X_test)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

print("AUC-ROC Score:", roc_auc_score(y_test, y_proba, multi_class='ovr'))

cv_scores = cross_val_score(dt_entropy, X, y_encoded, cv=5)
print("\nCross-Validation Scores:", cv_scores)
print("Mean CV Accuracy:", np.mean(cv_scores))
print("Train Accuracy:", dt_entropy.score(X_train, y_train))
print("Test Accuracy:", dt_entropy.score(X_test, y_test))
```
